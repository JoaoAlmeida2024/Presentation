```python
import statsmodel.api as sm
```

```python
# apply linear regression on the following data

Y = df1['Anxiety disorders']
X = df1.drop(['Anxiety disorders','Entity'], axis=1)
# The linear model has the formula Y = a * X + b, with 'add_constant' we force the model
# to have an intercept because by default on statsmodel library the `linear models`
# doesn't have intercept!!!
X = sm.add_constant(X)
model = sm.OLS(Y,X).fit()

model.summary()
```

```python
import pandas as pd
import pycountry
# Create an empty DataFrame
countries_df = pd.DataFrame()
# Collect data for each country
data = []
for country in pycountry.countries:
    try:
        current = {
            'alpha_3': country.alpha_3,
            'name': country.name,
            'official_name': country.official_name
        }
    except Exception:
        current = {
            'alpha_3': country.alpha_3,
            'name': country.name,
            'official_name': None
        }
    data.append(current)
# Populate the DataFrame
countries_df = pd.DataFrame(data, columns=['alpha_3', 'name', 'official_name'])
# Print the first few rows
countries_df
```
